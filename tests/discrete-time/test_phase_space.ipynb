```python
import numpy as np
import matplotlib.pyplot as plt
from pycandy import DiscreteDynamicalSystem as dds
```

```python
num_ic = 200
x_range = (0.0, 1)
y_range = (0., 1)


k = 1.5
u = np.random.rand(num_ic, 2)
u[:, 0] = x_range[0] + u[:, 0] * (x_range[1] - x_range[0])
u[:, 1] = y_range[0] + u[:, 1] * (y_range[1] - y_range[0])
```

```python
ds = dds(model="standard map")
```

```python
ts = ds.trajectory(u, 50000, parameters=k)
```

```python
plt.plot(ts[:, 0], ts[:, 1], "ko", markersize=0.1, markeredgewidth=0)
plt.xlim(x_range)
plt.ylim(y_range)
```
