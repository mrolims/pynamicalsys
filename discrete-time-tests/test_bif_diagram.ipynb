```python
from pycandy import DiscreteDynamicalSystem as dds
from pycandy import PlotStyler
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = np.array([0.1, 0.1])
parameters = np.array([0.3])
ds = dds(model="henon map")
```

```python
ds.info
```

```python
param_values, bd = ds.bifurcation_diagram(u, 0, (1, 1.4, 500), 10000, parameters=parameters, transient_time=5000)
```

```python
bd.shape
```

```python
ps = PlotStyler(markersize=0.1, markeredgewidth=0)
ps.apply_style()
for i in range(len(param_values)):
    plt.plot(param_values[i] * np.ones_like(bd[0]), bd[i], 'ko')
```
